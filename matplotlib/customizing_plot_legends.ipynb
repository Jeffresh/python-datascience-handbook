# Customizing Plot Legends

Plots legends give meaning to the visualization, assigning meaning to the various plot elements. We previoysly saw hot to create a simple legend; here we'll take a look at customizing the placemnet and aesthetics of the legend in Matplotlib.

The simplest legend can be created with the `plt.legend()` command, which automatically crates a legend for any labeled plot elements:

```python
import matplotlib.pyplot as plt 
plt.style.use('classic')
```

```python
%matplotlib inline 
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots() 
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend();
```

But there are many ways we ight want to customize such a legend. For example, we can specify the location and turn off the frame:

```python
ax.legend(loc='upper left', frameon=False)
fig
```

We can use the `ncol` command to specify the number of columns in the legend:

```python
ax.legend(frameon=False, loc='lower center', ncol=2)
fig
```

We can use a rounded box (`fancybox`) or add a shadow, change the transparency (alpha value) of the frame, or change the padding around the text:

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
```
